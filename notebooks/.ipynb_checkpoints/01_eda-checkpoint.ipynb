```python
# 01_eda.ipynb

# Importing the necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# For notebook visuals
%matplotlib inline
sns.set(style="whitegrid")

# Loading the dataset
df = pd.read_csv("../data/imdb_top_1000.csv")

# Preview
df.head()
```

## Dataset Shape and Info

```python
print(df.shape)
print(df.info())
print(df.describe())
```

## Check for Missing Values

```python
df.isnull().sum().sort_values(ascending=False)
```

## Unique Values

```python
df['Genre'].value_counts().head(10).plot(kind='barh', title='Top 10 Movie Genres')
plt.xlabel("Count")
plt.show()
```

## Cleaning the Data

```python
# Clean 'Gross' column (Revenue)
df['Gross'] = df['Gross'].str.replace(',', '')
df['Gross'] = pd.to_numeric(df['Gross'], errors='coerce')

# Clean 'Runtime' by removing 'min' and convert to int
df['Runtime'] = df['Runtime'].str.replace(' min', '')
df['Runtime'] = pd.to_numeric(df['Runtime'], errors='coerce')

# Drop rows where essential fields are missing
df.dropna(subset=['Gross', 'Runtime', 'IMDB_Rating'], inplace=True)

# Fill non-critical missing values
df['Meta_score'] = df['Meta_score'].fillna(df['Meta_score'].median())

# Convert 'Released_Year' to integer
df['Released_Year'] = pd.to_numeric(df['Released_Year'], errors='coerce')

# Reset the index
df.reset_index(drop=True, inplace=True)
```

## Saving the Cleaned Data

```python
df.to_csv("../data/imdb_cleaned.csv", index=False)
```

# Exploratory Data Analysis (EDA)

```python
# Distribution of IMDb Ratings
sns.histplot(df['IMDB_Rating'], kde=True, bins=20)
plt.title("Distribution of IMDb Ratings")
plt.xlabel("IMDB Rating")
plt.ylabel("Count")
plt.show()
```

```python
# Most Common Genres
df['Genre'].value_counts().head(10).plot(kind='barh', color='teal')
plt.title("Top 10 Movie Genres")
plt.xlabel("Number of Movies")
plt.gca().invert_yaxis()
plt.savefig('../images/top10_movie_genres.png')
plt.show()
```

```python
# Runtime vs. IMDb Rating
sns.scatterplot(data=df, x='Runtime', y='IMDB_Rating')
plt.title("Runtime vs IMDb Rating")
plt.xlabel("Runtime (minutes)")
plt.ylabel("IMDb Rating")
plt.savefig('../images/runtime_vs_rating.png')
plt.show()
```

```python
# Revenue vs. IMDb Rating

# Replace zeros or negative values (if any) before log transform
df = df[df['Gross'] > 0]

# Add a new column for log-transformed Gross
df['Log_Gross'] = np.log10(df['Gross'])

# Plot
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Log_Gross', y='IMDB_Rating', alpha=0.8)
plt.title("Log(Gross Revenue) vs IMDb Rating")
plt.xlabel("Log10(Gross Revenue)")
plt.ylabel("IMDb Rating")
plt.grid(True)
plt.savefig('../images/revenue_vs_rating.png')
plt.show()
```

```python
# Heatmap of Correlations
numeric_df = df[['IMDB_Rating', 'Runtime', 'Meta_score', 'Gross']]
corr = numeric_df.corr()

sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap")
plt.savefig('../images/correlation_heatmap.png')
plt.show()
```
