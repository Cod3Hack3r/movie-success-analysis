```python
# 01_eda.ipynb

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# For notebook visuals
%matplotlib inline
sns.set(style="whitegrid")

# Loading the dataset
df = pd.read_csv("../data/imdb_top_1000.csv")

df.head()
```

# Dataset Shape and Info

```python
print(df.shape)
print(df.info())
print(df.describe())
```

# Check for Missing Values

```python
df.isnull().sum().sort_values(ascending=False)
```

# Unique Values

```python
df['Genre'].value_counts().head(10).plot(kind='barh', title='Top 10 Movie Genres')
plt.xlabel("Count")
plt.show()
```
